```python
from shutil import copyfile
import glob
import csv
import pandas as pd
import json
import numpy as np
```

```python
top_dir = '.'
img_category = 'cars'
remap_path = '{0}/cars_mapping.xls'.format(top_dir)
remap_pd = pd.read_excel(remap_path, names=["new_name", "old_name", "correct key"], index_col=False)
```

### Generate list of lists that contain the key-mapping for each of the four parts of the experiment

```python
parts_list = ["prac", "part01", "part02", "part03"]
out_list = '[ '

for p, part in enumerate(parts_list):
    
    sub_pd = remap_pd[remap_pd["new_name"].str.contains(part)]
    
    # generate key mapping list
    key_list = sub_pd["correct key"].dropna().to_list()

    
    out_list += "["
    if p < 2:
        out_list += " ["
        images_per_id = 3
    for k, key in enumerate(key_list):
        out_list += str(int(key))
        if p >= 2:
            if (k+1 < len(key_list)):
                out_list += ","
            continue
        if (np.mod(k+1, images_per_id) == 0) and (k+1 < len(key_list)):
            # close the set
            out_list += "], ["
        elif (k+1 == len(key_list)):
            # close the group
            out_list += "] "
        else:
            out_list += ","
    if (p+1 < len(parts_list)):
        out_list += "], "
    else:
        out_list += "]"  
out_list += " ]"
print("copy the list below into your setup file:\n")
print(out_list)
print("\n")
```

### Rename files according to information in Excel file

```python
dry_run = True
old_dir = '{0}/{1}'.format(top_dir, img_category)
new_dir = '{0}/ready'.format(top_dir)

old_suffix = 'jpg'
new_suffix = 'jpg'

old_list = remap_pd["old_name"].dropna().to_list()
new_list = remap_pd["new_name"].dropna().to_list()

for n, old in enumerate(old_list):
        old_path = "{0}/{1}.{2}".format(old_dir, old, old_suffix)
        new_path = "{0}/{1}_{2}.{3}".format(new_dir, img_category, new_list[n], new_suffix)
        if not dry_run:
            copyfile(old_path, new_path)
        print(old_path.split("/")[-1], new_path.split("/")[-1])
```
